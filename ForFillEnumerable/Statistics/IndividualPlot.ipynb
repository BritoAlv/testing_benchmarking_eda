# Plot Graph of Our Data.

```r
library(ggplot2)
library(dplyr, warn.conflicts=FALSE)
library(repr)
options(repr.plot.width = 9, repr.plot.height = 5 ,repr.plot.res = 300)
```

# To do a  plot of  a Single DataFrame:

```r
#load the data
data <- read.csv(file = "./For.csv")
data
```

```r
# define the plot
x <- data$input_size
y  <- data$Tiempo
myplot <- ggplot(
  data = data, mapping = aes(x = x, y = y, colour = Algoritmo)) +
  geom_point(size = 2) +
    labs(x = "Tamaño De Entrada", y = "Tiempo Consumido (nanosegundos)") +
    geom_line(linetype = 2) +
    scale_x_continuous(breaks = x, labels = x) +
    theme(axis.text.x = element_text(angle = 90))
myplot  
```

```r
# collect some data to plot a linear function
cte_tiempo_best <- as.double(data$cte_tiempo[nrow(data)])
legend_curve <- paste(round(cte_tiempo_best, 6), "*x")
```

```r
myplot  <- myplot + stat_function(fun =  function(x) cte_tiempo_best*x, aes(colour = legend_curve))
myplot
```
